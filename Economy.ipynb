# Экономическая модель (в разработке)

## Виды валюты.
<b>Хард валюта</b> - ETH. Заработать напрямую в игре невозможно, должна использоватся для микротранзакций и как один из способов получения софт валюты.<br>
<b>Софт валюта</b> - RoboCoins. Заработать можно напрямую в игре, используется в ней повсеместно. Может косвенно переводится в хард валюту<b>(?)</b>.<br>
<br>
Для создания успешной и максимально прибыльной экономики требуется именно две валюты. Софт валюта предназначена для того, чтобы пользователь мог играть без вложений денег, хоть и с меньшей эффективностью чем игрок вложившийся. Хард валюта предназначена для продвинутых игроков которые уже вовлечены в игровой процесс("зацеплены"). С помощью хард валюты можно покупать игровую землю на которой можно поставить "бизнес".<br>
<br>
В число бизнесов входят:<br>
<ul>
    <li>Ферма - бизнес производящий MM(Metal Meat) и MV(Metal Veggies). Одна ферма может производить только один ресурс и после постройки не может переключиться на другой.</li>
    <br>
    <li>Шахта - бизнес производящий Нефть, MS(Metal Scrap) и Уголь. Одна шахта может производить только один ресурс и после постройки не может переключиться на другой.</li>
    <br>
    <li>Фабрика - бизнес перерабатывающий MS -> Одежда, MS -> Транспорт, MM и MS -> BS(Bolts and Screws) </li>
    <br>
    <li>Магазин - бизнес реализующий/перерабатывающий. Перерабатывает BS -> Продукты для чаров<b>(?)</b>, <b>(-)</b> -> Продукты для питомцев<b>(?)</b>.Реализует Продукты для чаров ии Продукты для питомцев по установленным владельцем ценам. </li>
    <br>
    <li>Заправка - перерабатывающий/реализующий. Перерабатывает Нефть -> Бензин. Реализует бензин по установленным владельцем ценамю</li>
    <br>
    <li>Бюро дальнобойщиков - реализующий/производящий. Перевозит ресурсы от одного бизнеса к другому. Использует бензин и транспорт.</li>
</ul>
Разные бизнесы производят и перерабатывают разные игровые ресурсы. Всего бизнесы разделяются на три вида: Производящие, Перерабатывающие, Реализующие.<br>

# Карта и покупка земли
Карта представлена в виде координатной сетки с зонами разных размеров (клетки), на которых можно построить "бизнес". При этом каждая клетка имеет свои характеристики, увеличивающие или уменьшающие производительность разных "бизнесов". Ключом к успеху является сопоставление характеристик клетки и соответствующего бизнеса. Клетки земли представлены в виде NFT токенов.

Игроки могут покупать незанятые клетки земли за хард валюту в любой момент. Стоимость земли высчитывается от её характеристик. Игроки могут покупать, продавать и передавать землю другим игрокам за софт и хард валюту.

# Бизнесы
Для постройки бизнеса на клетке требуется заплатить за него софт или хард валютой. Бизнес представлен в NFT токенах.<br>
